```python
import pandas as pd
import csv
import numpy
import random
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#STOCKS
```

```python
data = pd.read_csv('Assignment 1/q1_2012_all.csv')
stocks = []
for i in data[data['Security']=='Stock']['Ticker'].unique():
    stocks.append(i)
stocks = random.sample(stocks, 100)
print(len(stocks))
df = data[data['Ticker'].isin(stocks)]
df.reset_index(inplace=True)
df.drop(['index'],axis=1,inplace=True)
df.sort_values(by='Ticker',inplace=True)
df.reset_index(inplace=True)
df.drop(['index'],axis=1,inplace=True)
df
```

```python
df['Cancel_to_Trade'] = df['Cancels']/(df["OrderVol('000)"]*1000-df['Cancels'])
df['Trade-to-Order'] = (df["LitVol('000)"] + df["TradeVolForHidden('000)"] + df["TradeVolForOddLots('000)"])*100/(df["OrderVol('000)"])
df['Hidden_Rate'] = df['Hidden']*100/(df['LitTrades'] + df['OddLots'] + df['Hidden'])
df['Hidden_Vol'] = df["HiddenVol('000)"]*100/(df["LitVol('000)"] + df["TradeVolForHidden('000)"] + df["TradeVolForOddLots('000)"])
df['Oddlot_Rate'] = df['OddLots']*100/(df['LitTrades'] + df['TradesForHidden'] + df['TradesForOddLots'])
df['Oddlot_Vol'] = df["OddLotVol('000)"]*100/(df["LitVol('000)"] + df["TradeVolForHidden('000)"] + df["TradeVolForOddLots('000)"])
df.fillna(0, inplace=True)
df
```

```python
columns_of_interest = ['Cancel_to_Trade', 'Trade-to-Order', 'Hidden_Rate', 'Hidden_Vol', 'Oddlot_Rate', 'Oddlot_Vol']
stats = {}
for column in columns_of_interest:
    stats[column] = {
        'Mean': 0,
        'p25': 0,
        'p50': 0,
        'p75': 0,
        'SD': 0
    }
num = 0
for stock in stocks:
    df_stock = df[df['Ticker']==stock]
    print(len(df_stock))
    num+=1

    for column in columns_of_interest:
        stats[column]['Mean']+=df_stock[column].mean()
        stats[column]['p25']+=df_stock[column].quantile(0.25)
        stats[column]['p50']+=df_stock[column].quantile(0.50)
        stats[column]['p75']+=df_stock[column].quantile(0.75)
        stats[column]['SD']+=df_stock[column].std()
        if df_stock[column].std() == None:
            print(stock)
for column in columns_of_interest:
    stats[column]['Mean']=stats[column]['Mean']/num
    stats[column]['p25']=stats[column]['p25']/num
    stats[column]['p50']=stats[column]['p50']/num
    stats[column]['p75']=stats[column]['p75']/num
    stats[column]['SD']=stats[column]['SD']/num

stats = pd.DataFrame(stats).T
stock_stats = stats
stock_stats
```

```python
all_stock_stats = {}
for stock in stocks:
    df_stock = df[df['Ticker']==stock]
    columns_of_interest = ['Cancel_to_Trade', 'Trade-to-Order', 'Hidden_Rate', 'Hidden_Vol', 'Oddlot_Rate', 'Oddlot_Vol']

    stats = {}
    

    for column in columns_of_interest:
        
        stats[column] = {
            'Mean': df_stock[column].mean(),
            'p25': df_stock[column].quantile(0.25),
            'p50': df_stock[column].quantile(0.50),
            'p75': df_stock[column].quantile(0.75),
            'SD': df_stock[column].std()
        }
    
    stats_df = pd.DataFrame(stats).T
    all_stock_stats[stock] = stats_df
all_stock_stats
```
